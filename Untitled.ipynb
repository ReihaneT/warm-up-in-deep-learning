```python
import numpy as np 
import csv
import pandas as pd
import os
import matplotlib.pyplot as plt
import PIL
from pandas.io.parsers import read_csv

np.set_printoptions(threshold=np.nan)

class Face():
    def __init__(self,data,firstSizeX,firstSizeY,SecondSizeX,SecondSizeY):
        self._data=data
        self._dim1x=firstSizeX
        self._dim1y=firstSizeY
        self._dim2x=SecondSizeX
        self._dim2y=SecondSizeY
        self._convert()  
        
    def _convert(self):
        reshape_data=np.reshape(self._data,(self._dim1x,self._dim1y))
        img1=PIL.Image.fromarray(np.uint8(reshape_data))
        self._img=img1.resize((self._dim2x,self._dim2y))
        y=np.array( self._img ) 
        vecImg = y.astype( np.float32 )/255.0
        self._target_vector=np.reshape(vecImg,(self._dim2x*self._dim2y,))
       
        
    def getDim(self):
        return (self._dim2x*self._dim2y)
    
    def getData(self):
        return self._target_vector
    
    def showImage(self):
        plt.matshow(self._img)
        
class Eigenfaces():
    def __init__(self,faces):
        self._T=np.zeros((faces[1].getDim(),len(faces)))
        i=0
        for face in faces:
            self._T[:,i]=face.getData()
            i=i+1
            
    def calPsi(self):
        x,y=self._T.shape
        self._Psi=np.zeros((x,1))
        for row in range(0,x):
            d=0
            for col in range(0,y):
                d=d+self._T[row,col]
            self._Psi[row,0]=d
            
        self._Psi=self._Psi/y
        self._Psi=np.reshape(self._Psi,(x,1))
        # second method
        self._Psi2=np.sum(self._T,axis= 1)/y
        
    def calPhi(self):
        x,y=self._T.shape
        self._Psi=np.reshape(self._Psi,(x,1))
        self._Psi2=np.reshape(self._Psi2,(x,1))
        self._Phi=np.zeros((x,y))
        self._Phi2=np.zeros((x,y))
        for item in range(0,y):
            self._Phi[:,item]=self._T[:,item]-self._Psi[:,0]
        # second method
        self._Phi2=self._T - self._Psi2
        
    def calEigenVector(self):
        x,y=self._T.shape
        self._C=C=np.dot(self._Phi,self._Phi.T)/y
        self._C2=np.dot(self._Phi2,self._Phi2.T)/y
        self._w,self._v=np.linalg.eig( self._C )
        assert np.max( np.abs( np.imag(self._w) ) ) < 1e-10, "Eigenvalues are not real"
        self._w = np.real(self._w)
        self._v = np.real(self._v)
        order = np.argsort( self._w )[::-1]
        self._v = self._v[:,order]
        #second method
        self._w2,self._v2=np.linalg.eig( self._C )
        assert np.max( np.abs( np.imag(self._w2) ) ) < 1e-10, "Eigenvalues2 are not real"
        self._w2 = np.real(self._w2)
        self._v2 = np.real(self._v2)
        order = np.argsort( self._w2 )[::-1]
        self._v2 = self._v2[:,order]
        
    def displayEigenVec(self):
       
        eigmat=self._v[:,0].reshape((32,32))
        plt.matshow(eigmat)
        
        eigmat2=self._v2[:,0].reshape((32,32))
        plt.matshow(eigmat2)
        
        
# load data
train_data = csv.reader(open('training.csv', 'r'))
fname = 'training.csv'
df = read_csv(os.path.expanduser(fname)) 
df['Image'] = df['Image'].apply(lambda im: np.fromstring(im, sep=' '))
X = np.vstack(df['Image'].values)  
# array of Faces
X=X[:1000,:]
x_size, y_size=X.shape
faces=[]

for face in range(0,x_size-1):
    faces.append(Face(X[face,:],96,96,32,32))
    
# for obj in faces:
#     obj.showImage()

EF=Eigenfaces(faces)
EF.calPsi()
EF.calPhi()
EF.calEigenVector()
EF.displayEigenVec()
```
